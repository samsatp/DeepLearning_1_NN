```python
import numpy as np
import matplotlib.pyplot as plt
```

We can estimate gradient of a function by using its definition:

$$
\frac{\partial}{\partial x}f(x) = \lim_{h \rightarrow 0} \frac{f(x+h) - f(x)}{h}
$$

```python
f = lambda x: x**2 + 2*x + 1
xs = np.linspace(-10,10,100)
ys = [f(x) for x in xs]
plt.plot(xs, ys);
```

```python
def df(x):
    h = 0.00001
    return (f(x+h) - f(x))/h

print(f"df/dx at x = 5, {df(5):.4f}")
print(f"df/dx at x = -3, {df(-3):.4f}")
```
